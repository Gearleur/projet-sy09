```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../dataset/data_cleaned.csv')
df.dtypes
```

# RÃ©partion des Diagnosis

```python
sns.countplot(data=df, x="Diagnosis")
plt.show()
```

```python
plt.pie(df['Diagnosis'].value_counts(), labels=df['Diagnosis'].value_counts().index, autopct='%1.1f%%')
plt.title('RÃ©partition de Diagnosis')
plt.show()
```

# Variables quantitatives

```python
def visualisation_quantitative(var):
    fig, axes = plt.subplots(3, 2, figsize=(15, 10))

    # Boxplot
    sns.boxplot(data=df, x=var, ax=axes[0][0])
    sns.boxplot(data=df, x="Diagnosis", ax=axes[0][1], y=var)

    # Histogram
    sns.histplot(data=df, x=var, ax=axes[1][0])
    sns.histplot(data=df, x=var, kde=True, hue="Diagnosis", ax=axes[1][1])

    # Scatter plot
    sns.scatterplot(data=df, x=var, y="Diagnosis", hue="Diagnosis", ax=axes[2][0])
    
    # Bar plot
    sns.barplot(data=df, y=var, x="Diagnosis", hue="Diagnosis", ax=axes[2][1])

    plt.show()
```

```python
var_quantitatives = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
for var in var_quantitatives:
    print(var)
    visualisation_quantitative(var)
```
