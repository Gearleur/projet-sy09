# Variables quantitatives

```python
def visualisation_quantitative(var):
    fig, axes = plt.subplots(3, 2, figsize=(12, 8))

    # Boxplot
    sns.boxplot(data=df, x=var, ax=axes[0][0])
    sns.boxplot(data=df, x="Attrition", ax=axes[0][1], y=var)

    # Histogram
    sns.histplot(data=df, x=var, ax=axes[1][0])
    sns.histplot(data=df, x=var, kde=True, hue="Attrition", ax=axes[1][1])

    # Scatter plot
    sns.scatterplot(data=df, x=var, y="Attrition", hue="Attrition", ax=axes[2][0])
    
    # Bar plot
    sns.barplot(data=df, y=var, x="Attrition", hue="Attrition", ax=axes[2][1])

    plt.show()
```

```python
var_quantitatives = df.select_dtypes(include='int64').columns.tolist()
for var in var_quantitatives:
    print(var)
    visualisation_quantitative(var)
```

# Variables qualitatives

```python
def visualisation_qualitative(var):
    fig, axes = plt.subplots(1, 2, figsize=(15, 6))
    
    # Countplot
    sns.countplot(data=df, x=var, ax=axes[0])
    sns.countplot(data=df, hue="Attrition", ax=axes[1], x=var)

    plt.show()
```

```python
for var in df.columns.tolist():
    if var not in var_quantitatives and var != "Attrition":
        print(var)
        visualisation_qualitative(var)
```
